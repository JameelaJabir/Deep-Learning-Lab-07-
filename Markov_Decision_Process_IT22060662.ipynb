Create the following grid world.

**States:** A 10x10 grid

**Actions:** Up, Down, Left, Right

**Tranistion probabilities:**
* 0.7 in the direction of action
* 0.1 in the three other directions
* The robot bounces back to the same state near edges

**Rewards:**
* (7,8) has a reward +10
* (2,7) has a reward +3
* (4,3) has a reward -5
* (7,3) has a reward -10
* No reward in other states

This example is based on Decision Making Under Uncertainty by M.J. Kochenderfer.

```python
#Let's define MDP paras
def createGrid10World():
    def xy2s(y, x):
        x = max(x, 0)
        y = max(y, 0)
        x = min(x, 9)
        y = min(y, 9)
        out = np.ravel_multi_index(np.array([x,y]), (10,10))
        return out

    def s2xy(s):
        x, y = np.unravel_index(s, (10,10))
        return y, x

    def gridPlot(ax, im, title='', cmap='Blues'):
        pl.imshow(im, interpolation='none', cmap=cmap, origin='lower')
        pl.colorbar()
        ax.set_xticks(np.arange(0, 10, 1));
        ax.set_yticks(np.arange(0, 10, 1));
        ax.set_xticklabels(np.arange(0, 10, 1));
        ax.set_yticklabels(np.arange(0, 10, 1));
        ax.set_xticks(np.arange(-.5, 10, 1), minor=True);
        ax.set_yticks(np.arange(-.5, 10, 1), minor=True);
        ax.grid(which='minor', color='w', linestyle='-', linewidth=1)
        pl.title(title);
        return

    A = ['left', 'right', 'up', 'down']
    S = np.arange(100)
    T = np.zeros((len(S), len(A), len(S)))
    R = np.zeros((len(S), len(A)))
    for s in S:
        x, y = s2xy(s)
        if x == 2 and y == 7:
            R[s, :] = 3
        elif x == 7 and y == 8:
            R[s, :] = 10
        else:
            if x == 7 and y == 3:
                R[s, :] = -10
            elif x == 4 and y == 3:
                R[s, :] = -5
            elif x == 0:
                if y == 0 or y == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 0] = -0.7
            elif x == 9:
                if y == 0 or y == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 1] = -0.7
            elif y == 0:
                if x == 0 or x == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 3] = -0.7
            elif y == 9:
                if x == 0 or x == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 2] = -0.7

            for a in A:
                if a == 'left':
                    T[s, 0, xy2s(x - 1, y)] += 0.7
                    T[s, 0, xy2s(x + 1, y)] += 0.1
                    T[s, 0, xy2s(x, y - 1)] += 0.1
                    T[s, 0, xy2s(x, y + 1)] += 0.1
                elif a == 'right':
                    T[s, 1, xy2s(x + 1, y)] += 0.7
                    T[s, 1, xy2s(x - 1, y)] += 0.1
                    T[s, 1, xy2s(x, y - 1)] += 0.1
                    T[s, 1, xy2s(x, y + 1)] += 0.1
                elif a == 'up':
                    T[s, 2, xy2s(x, y + 1)] += 0.7
                    T[s, 2, xy2s(x, y - 1)] += 0.1
                    T[s, 2, xy2s(x - 1, y)] += 0.1
                    T[s, 2, xy2s(x + 1, y)] += 0.1
                elif a == 'down':
                    T[s, 3, xy2s(x, y - 1)] += 0.7
                    T[s, 3, xy2s(x, y + 1)] += 0.1
                    T[s, 3, xy2s(x - 1, y)] += 0.1
                    T[s, 3, xy2s(x + 1, y)] += 0.1

    for a, c_x, c_y in [(0,0,0), (0,0,9),(1,9,0),(1,9,9),(2,0,9),(2,9,9),(3,0,0),(3,9,0)]:
        R[xy2s(c_x,c_y),a] = -0.8

    discount = 0.9

    nextStates = {}
    for si in range(len(S)):
        for ai in range(len(A)):
            nextStates[(si,ai)] = np.where((T[si, ai, :] != 0) == True)[0]

    return {'S':S, 'A':A, 'T':T, 'R':R, 'discount':discount, 'nextStates':nextStates, 'gridPlot':gridPlot, 'xy2s':xy2s, 's2xy':s2xy}
```

```python
class MDP():
    def __init__(self):
        pass

    def inbuilt_init(self, mdp_dict):
        self.S = mdp_dict['S']
        self.A = mdp_dict['A']
        self.T = mdp_dict['T']
        self.R = mdp_dict['R']
        self.discount = mdp_dict['discount']
        self.nextStates = mdp_dict['nextStates']
        self.gridPlot = mdp_dict['gridPlot']
        self.xy2s = mdp_dict['xy2s']
        self.s2xy = mdp_dict['s2xy']
```

```python
#Define the MDP
mdp = MDP()
mdp.inbuilt_init(mdp_dict=createGrid10World())

#Plot states
pl.figure(figsize=(3,3))
mdp.gridPlot(ax=pl.gca(), im=mdp.S.reshape((10,10)), title='States', cmap='Greys')

#Plot rewards
pl.figure(figsize=(15,3))
pl.suptitle('Rewards', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=mdp.R[:,a].reshape((10,10)), title='a='+mdp.A[a], cmap='jet')
pl.show()

#Plot rewards
pl.figure(figsize=(15,3))
pl.suptitle('Rewards - clipped larger +ve values for visualizing the edges', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=np.clip(mdp.R[:,a].reshape((10,10)), -1, -0.5), title='a='+mdp.A[a], cmap='jet')
pl.show()

#Plot rewards
s0_x, s0_y = 3, 5
s0 = mdp.xy2s(s0_x, s0_y)
pl.figure(figsize=(15,3))
pl.suptitle('Transition probabilities T(s1|s0=({},{}),a)'.format(s0_x, s0_y), fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=mdp.T[s0,a,:].reshape((10,10)), title='a='+mdp.A[a], cmap='Blues')
pl.show()
```

### 1. Policy evaluation

Computing the utility, U.

$U^\pi_k(s) = R(s, \pi(s)) + \gamma \sum_{s'} T(s' \mid s, \pi(s))U^\pi_{k-1}(s')$

```python
def iterativePolicyEvaluation(mdp, policy, numIterations=10):
    U = np.zeros(len(mdp.S))
    U_old = copy.copy(U)
    for t in range(numIterations):
        U_old = copy.copy(U)
        for s in mdp.S:
            sum_term = sum(mdp.T[s, policy, ns] * U_old[ns] for ns in mdp.nextStates[(s, policy)])
            U[s] = mdp.R[s, policy] + mdp.discount * sum_term
    return U

numIterations = 5
pl.figure(figsize=(15,3))
pl.suptitle('Utilities', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    U = iterativePolicyEvaluation(mdp=mdp, policy=a, numIterations=numIterations)
    mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='a='+mdp.A[a], cmap='jet')
pl.show()
#print(np.round(U.reshape(10,10),3))
```

```python
def f(action, numIter=1):
    U = iterativePolicyEvaluation(mdp, policy=mdp.A.index(action), numIterations=numIter)
    pl.figure(figsize=(3,3))
    mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='Utility', cmap='jet')
    pl.show()

interactive_plot = interactive(f, action=widgets.ToggleButtons(options=['left', 'right', 'up', 'down']),
                               numIter=widgets.IntSlider(min=0, max=20, step=1, value=0))
interactive_plot
```

```python
#Value iteration
def valueIteration(mdp, numIterations=1):
    U = np.zeros(len(mdp.S))
    U_old = copy.copy(U)
    for t in range(numIterations):
        U_old = copy.copy(U)
        for s in mdp.S:
            max_value = -float('inf')
            for a in range(len(mdp.A)):
                value = mdp.R[s, a] + mdp.discount * sum(mdp.T[s, a, ns] * U_old[ns] for ns in mdp.nextStates[(s, a)])
                max_value = max(max_value, value)
            U[s] = max_value
    return U

def policyExtration(mdp, U):
    policy = np.zeros(len(mdp.S))
    for s in mdp.S:
      max_value = -float('inf')
      best_action = 0
      for a in range(len(mdp.A)):
          value = mdp.R[s, a] + mdp.discount * sum(mdp.T[s, a, ns] * U[ns] for ns in mdp.nextStates[(s, a)])
          if value > max_value:
              max_value = value
              best_action = a
      policy[s] = best_action
    return policy

U = valueIteration(mdp, numIterations=2)
policy = policyExtration(mdp, U=U)
pl.figure(figsize=(3,3))
mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='Utility', cmap='jet')
for s in range(100):
    x, y = mdp.s2xy(s)
    if policy[s] == 0:
        m='\u02C2'
    elif policy[s] == 1:
        m='\u02C3'
    elif policy[s] == 2:
        m='\u02C4'
    elif policy[s] == 3:
        m='\u02C5'
    pl.text(x-0.5,y-1,m,color='k',size=20)
pl.show()
```

### 2. Policy iteration

Policy evaluation can be used in policy iteration:
1. Given the current policy, compute U
2. Using U, compute a new policy

```python
def policyIteration(mdp, numIterations=1):
    U_pi_k = np.zeros(len(mdp.S))  # Initial values
    pi_k = np.random.randint(low=0, high=4, size=len(mdp.S), dtype=int)  # Initial policy
    pi_kp1 = copy.copy(pi_k)
    for t in range(numIterations):
        # Policy evaluation: compute U_pi_k
        U_old = copy.copy(U_pi_k)
        for s in mdp.S:
            sum_term = sum(mdp.T[s, pi_k[s], ns] * U_old[ns] for ns in mdp.nextStates[(s, pi_k[s])])
            U_pi_k[s] = mdp.R[s, pi_k[s]] + mdp.discount * sum_term
        # Policy improvement
        for s in mdp.S:
            max_value = -float('inf')
            best_action = pi_k[s]
            for a in range(len(mdp.A)):
                value = mdp.R[s, a] + mdp.discount * sum(mdp.T[s, a, ns] * U_pi_k[ns] for ns in mdp.nextStates[(s, a)])
                if value > max_value:
                    max_value = value
                    best_action = a
            pi_kp1[s] = best_action
        pi_k = copy.copy(pi_kp1)
    return U_pi_k, pi_kp1

U_pi_k, pi_kp1 = policyIteration(mdp, numIterations=2)
```

# **Model Free Approach**

Adding a Q Agent

```python
class QLearningAgentMDP:
    def __init__(self, mdp, epsilon=0.1, alpha=0.1, gamma=0.9):
        self.mdp = mdp
        self.q_table = np.zeros((len(mdp.S), len(mdp.A)))
        self.epsilon = epsilon
        self.alpha = alpha
        self.gamma = gamma

    def choose_action(self, state):
        # Epsilon-greedy action selection
        if np.random.rand() < self.epsilon:
            return np.random.choice(len(self.mdp.A))
        else:
            max_q = np.max(self.q_table[state])
            best_actions = np.where(self.q_table[state] == max_q)[0]
            return np.random.choice(best_actions)

    def learn(self, state, action, reward, next_state):
        # Q-Learning update
        self.q_table[state, action] = self.q_table[state, action] + self.alpha * (
            reward + self.gamma * np.max(self.q_table[next_state]) - self.q_table[state, action]
        )

    def train(self, episodes=500, max_steps=100):
        rewards_per_episode = []
        for ep in range(episodes):
            s = np.random.choice(self.mdp.S)  # random start state
            cumulative_reward = 0
            for step in range(max_steps):
                a = self.choose_action(s)
                next_states = self.mdp.nextStates[(s, a)]

                # Fix: handle empty next_states
                if len(next_states) == 0:
                    s1 = s  # stay in the same state
                else:
                    probs = self.mdp.T[s, a, next_states]
                    probs = probs / np.sum(probs)
                    s1 = np.random.choice(next_states, p=probs)

                r = self.mdp.R[s, a]
                self.learn(s, a, r, s1)
                cumulative_reward += r
                s = s1

            rewards_per_episode.append(cumulative_reward)
        return rewards_per_episode
```

Compare Execution Time

```python
import time
# Model-Based: Policy Iteration
start_time = time.time()
U_pi, pi = policyIteration(mdp, numIterations=10)
time_pi = time.time() - start_time
print("Policy Iteration time:", np.round(time_pi,3), "seconds")

# Model-Free: Q-Learning
agentQ = QLearningAgentMDP(mdp, epsilon=0.1, alpha=0.1, gamma=mdp.discount)
start_time = time.time()
rewards = agentQ.train(episodes=500)
time_q = time.time() - start_time
print("Q-Learning training time:", np.round(time_q,3), "seconds")
```

## Model-Based vs Model-Free Reinforcement Learning

1. Model-Based RL (e.g., Policy Iteration, Value Iteration)

In a model-based method, the agent knows the model of the environment.

This means it knows:

Transition probabilities T(s,a,s′) (what happens when you take an action).

Reward function R(s,a) (the immediate reward).

Using this knowledge, the agent can plan ahead by simulating possible future states.

Examples:

Policy Iteration

Alternates between evaluating a policy and improving it.

Value Iteration

Directly updates state values using the Bellman optimality equation until convergence.

Pros: Efficient if the environment is fully known.
Cons: Requires complete knowledge of the environment (not always possible in real-world problems).

2. Model-Free RL (e.g., Q-Learning, SARSA)

Definition: The agent learns only from interactions with the environment. No model of transitions or rewards is assumed.

How it works:

Uses experience (state, action, reward, next state) to update action-values or policies.

Explores the environment using epsilon-greedy or similar strategies.

Example: Q-Learning updates a Q-table based on observed rewards and the estimated future value.

Pros:

Can be applied to environments where the model is unknown.

Learns optimal or near-optimal policies through trial-and-error.

Cons:

Slower convergence, stochastic learning.

Requires many episodes to explore the environment sufficiently.
